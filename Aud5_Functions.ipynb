Функција е блок од наредби во код кој се извршува само кога е повикан.

Може да се испратат податоци во функција, познати како параметри.

Функцијата може да врати различни податочни типови како резултат.

Програмите претставуваат комбинација од кориснички и библиотечни функции


### Дефиниција и повикување на функција


Во Python, функција се дефинира со def клучниот збор:

```python
#DEFINITION
def my_function():
    print("Hello world, from a function")
    return 1
```

```python
result = my_function()
print(result)
```

```python
#DEFINITION
def ime_na_funkcija():
    print("Hello World!")
```

```python
#CALLS
my_function()
ime_na_funkcija()
```

### Аргументи
Информациите може да се пренесат во функции како аргументи.

Аргументите се наведени по името на функцијата, внатре во заградите. Можете да додавате колку сакате аргументи, само одвојте ги со запирка.

Следниот пример има функција со еден аргумент (fname). Кога се повикува функцијата, пренесуваме име, кое се користи внатре во функцијата за да се испечати целото име:

Променливите декларирани во листата на параметри важат и може да се употребат само во телото на функцијата 

```python
def funkcija(fname):
  print(fname + " IME")

funkcija("Emil")
funkcija("Aleksandar")
funkcija("Ana")
```

```python
def funkcija(fname):
    fname = "Ana"
    print(fname + " IME")

fname = "Darko"
funkcija(fname)
print(f"Fname nadvor od f-ja {fname}")
```

### Број на аргументи

- Стандардно, мора да се повика функција со точен број на аргументи.
- Ако вашата функција очекува 2 аргументи, треба да ја повикате функцијата со 2 аргументи, не повеќе, ниту помалку.

```python
def funkcija(fname, lname):
  print(fname + " " + lname)

funkcija("Aleksandar", "Stojmenski")
funkcija("Petar", "Petrovski")
```

```python
funkcija("Petar")
```

```python
funkcija("Petar", "Ivanov", "Petrov")
```

- На овој начин функцијата ќе добие неколку аргументи и ќе може соодветно да пристапите до нив:

### Област на важење на променливи



```python
vkupno = 0 # Ova e globalna promenliva
def zbir (x , y): # Definicija na funkcija
    vkupno = x + y # Ova e lokalna promenliva
    print("Vo funkcijata , vkupno = ", vkupno)
    return vkupno

zbir(10 , 20)
print("Nadvor od funkcijata , vkupno = ", vkupno)
```

Ако локална и глобална променлива имаат исто име, локалната променлива „фрла сенка на“ (ја препокрива) глобалната променлива

```python
vkupno = 0 # Ova e globalna promenliva
def zbir (x , y): # Definicija na funkcija
    global vkupno
    vkupno = x + y # Ova e lokalna promenliva
    print("Vo funkcijata , vkupno = ", vkupno)
    return vkupno

zbir(10 , 20)
print("Nadvor od funkcijata , vkupno = ", vkupno)
```

### Пример 7: Напишете функција која ќе користи while наредба за да ги собере сите броеви до 100 (вклучително).

```python
def suma_100():
    s = 0
    i = 1
    while i<=100:
        s+=i
        i+=1
    return s

print(suma_100())
```

### Пример 9: Да се напише код за игра во која програмата избира случаен број, а играчот треба да го погоди тој број. Играта дава совети, како што се „Високо“ или „Ниско“, додека играчот не погоди правилно. На крај испечатети ги бројот на обиди.

```python
import random

# Generate a random number between 1 and 100
def pogodi():
    secret_number = random.randint(1, 100)
    i = 0
    while True:
        n = int(input())
        i += 1
        if n== secret_number:
            print("Go pogodivte brojot")
            break
        elif n < secret_number:
            print("Nisko")
        elif n > secret_number:
            print("Visoko")

    return i

print(pogodi())
```
