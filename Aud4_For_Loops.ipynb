### Пример 1: Пресметај ја сумата на броевите од 1 до 10

````python
s=0

for i in range(1, 11):
    s+=i

print(s)
````

### Пример 4: Пресметување на факториел на одреден број.

````python
p = 1

n = int(input())
for i in range(1, n+1):
    p*=i

print(f'{n}! = {p}')
````

### Пример 5: Најди ги сите делители на број кој се внесува од тастатура

````python
n = int(input())

for i in range(1, n+1):
    if n%i == 0:
        print(i)
````

### Пример 6: Да се најдат сите прости броеви во опсег (почеток и крај кои се читаат од стандарден влез):


````python
n = int(input())

is_prost = True

for i in range(2, n):
    if n%i == 0:
        is_prost = False

if is_prost:
    print("Prost")
else:
    print("Ne e prost")
````

````python
n = int(input("Vnesi n"))
m = int(input("Vnesi m"))

for i in range(n, m+1):
    is_prost = True
    
    for j in range(2, i):
        if i%j == 0:
            is_prost = False
    
    if is_prost:
        print(i)
````

### Пример 7: Да се напише програма што од N цели броеви внесени од тастатура ќе го определи најголемиот број внесен на парна позиција (според редоследот на внесување) и најмалиот на непарна позиција (според редоследот на внесување)

````python
n = int(input("Vnesi n"))

maxi=0
mini=0

flag=0

for i in range(0, n):
    m = int(input("Vnesi broj"))

    if flag == 0:
        maxi = m
        mini = m
        flag = 1

    if i%2==0:
        if m >= maxi:
            maxi = m
    else:
        if m <= mini:
            mini = m

print(f'Maks e {maxi}')
print(f'Mini e {mini}')
````

### Пример 8: Да се внесат поени од тест за N студенти и да се пресмета просечниот број на поени освоени од тестовите на сите студенти.

````python
n = int(input("Vnesi broj na studenti"))

s = 0

for i in range(0, n):
    p = int(input(f"Vnesi poeni za {i+1}-tiot student"))

    s += p

print(f'Prosek = {s/n}')
````
