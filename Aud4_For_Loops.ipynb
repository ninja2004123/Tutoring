###  6:         (        ):


````python
n = int(input())

is_prost = True

for i in range(2, n):
    if n%i == 0:
        is_prost = False

if is_prost:
    print("Prost")
else:
    print("Ne e prost")
````

````python
n = int(input("Vnesi n"))
m = int(input("Vnesi m"))

for i in range(n, m+1):
    is_prost = True
    
    for j in range(2, i):
        if i%j == 0:
            is_prost = False
    
    if is_prost:
        print(i)
````

###  7:       N               (   )      (   )

````python
n = int(input("Vnesi n"))

maxi=0
mini=0

flag=0

for i in range(0, n):
    m = int(input("Vnesi broj"))

    if flag == 0:
        maxi = m
        mini = m
        flag = 1

    if i%2==0:
        if m >= maxi:
            maxi = m
    else:
        if m <= mini:
            mini = m

print(f'Maks e {maxi}')
print(f'Mini e {mini}')
````

###  8:        N               .

````python
n = int(input("Vnesi broj na studenti"))

s = 0

for i in range(0, n):
    p = int(input(f"Vnesi poeni za {i+1}-tiot student"))

    s += p

print(f'Prosek = {s/n}')
````
