Контролни структури за избор
(if...elif...else)

Most of the times, you want to control the flow of your program. There can be certain business rules or conditions your client or you want to satisfy in order to be successful. This means that you will want to execute one line or block of code when if a condition is satisfied and a different set of code if it is not.

The if-elif-else statement is used to conditionally execute a statement or a block of statements. Conditions can be true or false, execute one thing when the condition is true, something else when the condition is false.

### `Else`
The else keyword catches anything which isn't caught by the preceding conditions.

###### `if` (condition) :
      Indented statement block for when condition is TRUE
###### `else`:
      Indented statement block for when condition is FALSE

### `Elif`

###### `if` (condition) :
      Indented statement block for when condition is TRUE
###### `elif` (condition):
      Indented statement block for when condition is TRUE
###### `else`:
      Indented statement block for when condition is FALSE

### Пример6: Напиши програма која ќе проверува дали поените од теоретскиот и практичниот дел се до 50, потоа провери 3 различни ситуации:
- 	poeni_teorija = 60 , poeni_od_prakitcen_del = 20

- 	poeni_teorija = 30 , poeni_od_prakitcen_del = 66

- 	poeni_teorija = 52 , poeni_od_prakitcen_del = 76

Доколку поените од било кој дел се под 50 студентот треба да го преполага тој дел, ако збирот на поените е над 100, тогаш испринтај колку поени има освоени студентот од испитот.

````python
student = input("Vnesete student: ")
teo = int(input("Vnesi poeni teorija: "))
prak = int(input("Vnesi poeni prak: "))

if teo < 50 or prak < 50 or teo + prak <100:
    print(f"Studentot {student} e padnat")
else:
    print(f"Studentot {student} polozhil i osvoil {teo+prak}")
````

### Пример10: Внеси три вредности на варијабли преку input поле и со помош на услови испринтај го најголемиот број

### Nested if else

````python
predmet = input('Vnesete predmet, algebra ili geometrija: ')
teorija = 53
zadaci = 35
````

````python
if (predmet.lower() == 'algerba'):
    if teorija < 50:
        print('Treba da prepolaga teorija po ', predmet)
    elif zadaci < 50:
        print('Treba da prepolaga zadaci po ', predmet)
    else:
        print('Vie imate:', teorija + zadaci)
elif (predmet == 'geometrija' or predmet == 'Geometrija' or predmet == 'GEOMETRIJA'):
    if teorija < 50:
        print('Treba da prepolaga teorija po', predmet)
    elif zadaci < 50:
        print('Treba da prepolaga zadaci po', predmet)
    else:
        print('Vie imate:', teorija + zadaci)
else:
    print('Predmetot ne e pronajden')
````
